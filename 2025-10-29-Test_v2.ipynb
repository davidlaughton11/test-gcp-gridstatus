```python
from gridstatusio import GridStatusClient
import os
from dotenv import load_dotenv

load_dotenv()  # reads .env into os.environ

GRIDSTATUS_API_KEY = os.getenv("GRIDSTATUS_API_KEY")

client = GridStatusClient(api_key=GRIDSTATUS_API_KEY)

QUERY_LIMIT = 100
```

```python

START = "2025-10-10"
END = "2025-10-28"

data_PJM_apr = client.get_dataset(
    dataset="pjm_lmp_day_ahead_hourly",
    start=START,
    end=END,
    filter_column="location",
    filter_operator="in",
    filter_value=["DOMINION HUB",
        "WESTERN HUB",
        "EASTERN HUB",
        '02HAMILT138 KV  LD1',
        '02HAMLTN138 KV  TR1',
        '02HAMLTN138 KV  TR2',
        '02HAMLTN138 KV  TR3',
        '02HAMLTN138 KV  TR4',
        '02HUMMEL36 KV   LDFDR',
        '02INCA  138 KV  TR71',
        '02IRON  138 KV  TR71',
        '02IRON  138 KV  TR72',
        '02JMP   138 KV  TR1',
        '02JMP2  138 KV  TR10',
        '02JMP2  138 KV  TR11',
        '02JOHNCT138 KV  LDJC',
        '02JOHNCT138 KV  LDJC2',
        '02JOHNCT138 KV  TR1',
        '02JP    138 KV  TR1',
        '02JP    138 KV  TR2',
        '02JP    138 KV  TR3',
        '02KYHS  138 KV  TR1',
        '02KYHS  138 KV  TR2',],
    timezone="market",
    limit=QUERY_LIMIT,
)

data_PJM_apr
```

```python
import os
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "gcp-sa-key.json"

import pandas as pd
from google.cloud import storage
from io import BytesIO

def upload_parquet_to_gcs_in_memory(df, bucket_name, dest_blob):
    # Convert DataFrame to Parquet in memory
    buffer = BytesIO()
    df.to_parquet(buffer, index=False)
    buffer.seek(0)  # Move to the start of the buffer

    # Upload to GCS
    client = storage.Client()
    bucket = client.bucket(bucket_name)
    blob = bucket.blob(dest_blob)
    blob.upload_from_file(buffer, content_type='application/octet-stream')
    print(f"✅ Uploaded DataFrame to gs://{bucket_name}/{dest_blob}")

# Example usage
upload_parquet_to_gcs_in_memory(data_PJM_apr, "my-data-bucket-davidl", "results/pjm_LMP_data.parquet")
```

```python
# read back the data
df = pd.read_parquet("gs://my-data-bucket-davidl/results/pjm_LMP_data.parquet")
print(df.head())
```
